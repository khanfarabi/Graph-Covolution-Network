```python
# GCN Prediction and Explanations.


import pandas as pd
import os

import stellargraph as sg
from stellargraph.mapper import FullBatchNodeGenerator
from stellargraph.layer import GCN

from tensorflow.keras import layers, optimizers, losses, metrics, Model
from sklearn import preprocessing, model_selection
from IPython.display import display, HTML
import matplotlib.pyplot as plt
from tensorflow.keras.callbacks import EarlyStopping
from sklearn.metrics import f1_score
def GCN_Result():
        class gcn:
            @classmethod
            def data(cls,dataset):
                display(HTML(dataset.description))
                G, node_subjects = dataset.load()
                return G, node_subjects
            @classmethod
            def update_inp_mlev(cls,node_subjects,WORDS):
                rm1=[]
                for t in node_subjects.index:
                    #print(t)
                    if str(t) not in WORDS:
                        rm1.append(t)
                print(len(rm1))
                node_subjects1=node_subjects.drop(labels=rm1)
                return node_subjects1
            @classmethod
            def dis_graph_info(cls,G):
                return  G.info()
            @classmethod
            def updated_nodes(cls,node_subjects1):
                return node_subjects1.value_counts().to_frame()
            @classmethod
            def train_val(cls,node_subjects1):
                train_subjects, test_subjects = model_selection.train_test_split(node_subjects1, train_size=100, test_size=None, stratify=node_subjects1)
                val_subjects, test_subjects = model_selection.train_test_split(test_subjects, train_size=200, test_size=None, stratify=test_subjects)
                return train_subjects,val_subjects, test_subjects
            @classmethod
            def encoding(cls,train_subjects,val_subjects,test_subjects):
                target_encoding = preprocessing.LabelBinarizer()
                train_targets = target_encoding.fit_transform(train_subjects)
                val_targets = target_encoding.transform(val_subjects)
                test_targets = target_encoding.transform(test_subjects)
                return target_encoding,train_targets, val_targets,test_targets
            @classmethod
            def gcn_model(cls,G,target_encoding,train_targets, val_targets,test_targets,node_subjects1):
                generator = FullBatchNodeGenerator(G, method="gcn")
                train_gen = generator.flow(train_subjects.index, train_targets)
                gcn = GCN(layer_sizes=[16, 16], activations=["relu", "relu"], generator=generator, dropout=0.5)
                x_inp, x_out = gcn.in_out_tensors()
                predictions = layers.Dense(units=train_targets.shape[1], activation="softmax")(x_out)
                model = Model(inputs=x_inp, outputs=predictions)
                model.compile(optimizer=optimizers.Adam(lr=0.01),loss=losses.categorical_crossentropy,metrics=["acc"],)
                val_gen = generator.flow(val_subjects.index, val_targets)
                es_callback = EarlyStopping(monitor="val_acc", patience=50, restore_best_weights=True)
                history = model.fit(train_gen,epochs=200,validation_data=val_gen,verbose=2,shuffle=False,  callbacks=[es_callback],)
                # this should be False, since shuffling data means shuffling the whole graph
                sg.utils.plot_history(history)
                test_gen = generator.flow(test_subjects.index, test_targets)
                '''
                test_metrics = model.evaluate(test_gen)
                print("\nTest Set Metrics:")
                for name, val in zip(model.metrics_names, test_metrics):
                    print("\t{}: {:0.4f}".format(name, val))
                '''
                all_nodes = node_subjects1.index
                all_gen = generator.flow(all_nodes)
                all_predictions = model.predict(all_gen)
                node_predictions = target_encoding.inverse_transform(all_predictions.squeeze())
                #mapping categories to numbers
                d_tt1={}
                #margs2={}
                d_tt1["Neural_Networks"]=0
                d_tt1["Probabilistic_Methods"]=1
                d_tt1["Genetic_Algorithms"]=2
                d_tt1["Theory"]=3
                d_tt1["Case_Based"]=4
                d_tt1["Reinforcement_Learning"]=5
                d_tt1["Rule_Learning"]=6
                # mapping and computing f-score
                gcnp={}
                gcnr={}
                for t in range(0,len(node_predictions)):#node_predictions:
                    if str(node_subjects1.index[t]) in WORDS:
                        #print(node_subjects1.index[t],node_subjects1[node_subjects1.index[t]],node_predictions[t])
                        gcnp[node_subjects1.index[t]]=node_predictions[t]
                        gcnr[node_subjects1.index[t]]=node_subjects1[node_subjects1.index[t]]
                p=[]
                tr=[]
                for k in gcnp:
                    p.append(d_tt1[gcnp[k]])
                for k in gcnr:
                    #print(d_tt1[gcnr[k]])
                    tr.append(d_tt1[gcnr[k]])

                rs=f1_score(tr, p, average='micro')
                return rs,gcnp,x_inp, x_out,all_gen


        dataset = sg.datasets.Cora()
        G, node_subjects=gcn.data(dataset)
        node_subjects1=gcn.update_inp_mlev(node_subjects,WORDS)
        train_subjects,val_subjects, test_subjects=gcn.train_val(node_subjects1)
        target_encoding,train_targets, val_targets,test_targets=gcn.encoding(train_subjects,val_subjects,test_subjects)
        rs,gcnp,x_inp, x_out,all_gen=gcn.gcn_model(G,target_encoding,train_targets, val_targets,test_targets,node_subjects1)
        print(rs)


        #Embedding Representation
        def embed(x_inp,x_out,all_gen):
            embedding_model = Model(inputs=x_inp, outputs=x_out)
            emb = embedding_model.predict(all_gen)
            emb.shape
            from sklearn.decomposition import PCA
            from sklearn.manifold import TSNE

            transform = TSNE  # or PCA
            X = emb.squeeze(0)
            X.shape
            trans = transform(n_components=2)
            X_reduced = trans.fit_transform(X)
            X_reduced.shape
            fig, ax = plt.subplots(figsize=(7, 7))
            ax.scatter(
                X_reduced[:, 0],
                X_reduced[:, 1],
                c=node_subjects1.astype("category").cat.codes,
                cmap="jet",
                alpha=0.7,
            )
            ax.set(
                aspect="equal",
                xlabel="$X_1$",
                ylabel="$X_2$",
                title=f"{transform.__name__} visualization of GCN embeddings for cora dataset",
            )
        embed(x_inp,x_out,all_gen)
        #MLN explanation of the GCN Prediction Store_Explanation_r
        def gcn_exp(gcnp,Store_Explanation_r):
            for k in gcnp:
                if str(k) in Store_Explanation_r:
                    pass
                          #print(k,gcnp[k],Store_Explanation_r[str(k)])
                          #print("\n")



        gcn_exp(gcnp,Store_Explanation_r)

GCN_Result()       
```
